```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['education'] = df['education'].fillna(df['education'].mean())
df['cigsPerDay'] = df['cigsPerDay'].fillna(df['cigsPerDay'].mean())
df['BPMeds'] = df['BPMeds'].fillna(df['BPMeds'].mean())
df['BMI'] = df['BMI'].fillna(df['BMI'].mean())
df['heartRate'] = df['heartRate'].fillna(df['heartRate'].mean())
df['glucose'] =df['glucose'].fillna(df['glucose'].mean())
df['totChol'] =df['totChol'].fillna(df['totChol'].mean())
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
x =df[['age','prevalentHyp','sysBP','glucose','diabetes','diaBP']]
y= df['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=123456)
```

```python
def create_logistic_Reg_model():
    from sklearn.linear_model import LogisticRegressionCV
    model = LogisticRegressionCV()
    model.fit(x_train,y_train)
    return model
```

```python
def create_KNN_model():
    from sklearn.neighbors import KNeighborsClassifier
    model = KNeighborsClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
def create_SVM_model():
    from sklearn.svm import SVC
    model = SVC()
    model.fit(x_train,y_train)
    return model
```

```python
def create_naive_bayes_model():
    from sklearn.naive_bayes import GaussianNB
    model = GaussianNB()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
def evaluate_model(model):
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)
    return f'{accuracy:.2f}', f'{precision:.2f}', f'{recall:.2f}', f'{f1:.2f}'
    
```

```python
# list to keep all models
models =[]
#create all the models
models.append({"name" : "Logistic Regression",'model':create_logistic_Reg_model() })
models.append({"name" : "naive bayes",'model':create_naive_bayes_model() })
models.append({"name" : "KNN",'model':create_KNN_model()})
models.append({"name" : "SVM",'model':create_SVM_model()})
```

```python
for model in models:
    accuracy,precision,recall,f1=evaluate_model(model['model'])
    model['accuracy'] = accuracy
    model['precision'] = precision
    model['recall'] = recall
    model['f1'] = f1

df_evaluation_report_test = pd.DataFrame(models,columns=['name','accuracy','precision','recall','f1']) 
df_evaluation_report_test 

    
```
